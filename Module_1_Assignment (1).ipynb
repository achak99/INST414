Data Import and Cleaning


```python
import pandas as pd

# Load the dataset
data = pd.read_csv('downloads/digital_marketing_campaigns.csv')

# Handle missing cost values
data['Cost'] = data['Cost'].fillna(data['Cost'].median())

# Calculate acquisition rate
data['AcquisitionRate'] = data['Conversions'] / data['Impressions']

# Remove duplicates
data = data.drop_duplicates(subset='CampaignID')

print(data.head())
```

# Summary Statistics

```python
summary_stats = data.describe()
print(summary_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=data, x='Channel', y='AcquisitionRate', palette='viridis')
plt.title('Customer Acquisition Rate by Marketing Channel')
plt.show()
```

Analysis: plotting spend versus conversions to identify diminishing returns

```python
sns.scatterplot(data=data, x='Cost', y='Conversions', hue='Channel', palette='coolwarm')
plt.title('Campaign Spend vs Conversions')
plt.show()
```
